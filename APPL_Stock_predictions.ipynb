```python
import yfinance as yf
```

```python
APPL = yf.Ticker("AAPL")
```

```python
APPL = APPL.history(period="max")
```

```python
APPL.plot.line(y="Close", use_index=True)
```

```python
del APPL["Dividends"]
del APPL["Stock Splits"]
```

```python
APPL["TOMORROW"] = APPL["Close"].shift(-1)
```

```python
APPL
```

```python
APPL["Target"] = (APPL["TOMORROW"] > APPL["Close"]).astype(int)
```

```python
APPL
```

```python
APPL = APPL.loc["1990-01-01":].copy()
```

```python
APPL
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
train = APPL.iloc[:-100]
test = APPL.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(APPL, model, predictors)
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
APPL["SMA_5"] = APPL["Close"].rolling(5).mean()
APPL["SMA_10"] = APPL["Close"].rolling(10).mean()
APPL["SMA_20"] = APPL["Close"].rolling(20).mean()
```

```python
APPL["Momentum"] = (APPL["Close"] - APPL["Close"].shift(5))
```

```python
APPL["Volatility"] = APPL["Close"].rolling(5).std()
```

```python

def compute_rsi(Series, periods=14):
    delta = Series.diff()
    gain = delta.where(delta > 0, 0).rolling(window=periods).mean()
    loss = -delta.where(delta < 0, 0).rolling(window=periods).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))
APPL["RSI"] = compute_rsi(APPL["Close"])
```

```python
APPL["Lag_1"] = APPL["Close"].shift(1)
APPL["Lag_2"] = APPL["Close"].shift(2)
```

```python
predictors = ["SMA_5", "SMA_10", "SMA_20", "Momentum", "Volatility", "RSI", "Lag_1", "Lag_2"]
```

```python
predictions = backtest(APPL, model, predictors)
precision_score(predictions["Target"], predictions["Predictions"])
```
